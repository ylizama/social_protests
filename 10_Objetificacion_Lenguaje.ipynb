```python
%reload_ext cypher
import pandas as pd
```

## Threads - Conversaciones

```python
all_paths_df =  pd.DataFrame(columns=['tweet_id', 'path', 'path_len', 'relationships', 'hours_diff', 'ids']) 
```

```python
with open('conversations/ids.txt') as f:
    tweets_list = f.read().splitlines()
tweets_list = tweets_list[91364:]
```

```python
i = len(tweets_list)
print(i)
```

```python
for tweet_id in tweets_list:
    df = %cypher MATCH path=(n:tweet)-[r :RETWEETS|REPLIES_TO|QUOTED_TO*]-> (m:tweet) WHERE n.tid = '{tweet_id}' RETURN n.tid as tweet_id, path, length(path) as path_len, [r in relationships(path) | type(r)] as relationships, [t in nodes(path) | t.tid] as nodes ORDER BY path_len DESC LIMIT 1
    df = df.get_dataframe()
    all_paths_df = pd.concat([all_paths_df, df], ignore_index=True, sort=False)
    i = i-1
    print(i)
```

```python
len(all_paths_df)
```

```python
all_paths_df.to_csv("conversations/h9.csv", index=False, header=False, sep="\t")
```
